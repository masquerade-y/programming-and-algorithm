# Linked List Reversal 

## Problem

Write a function to reverse a Linked List in place. The function will take in the head of the list as input and return the new head of the list.

You are given the example Linked List Node class:

```python
class Node(object):
    
    def __init__(self,value):
        
        self.value = value
        self.nextnode = None
```

# Solution

Fill out your solution below:

```python
# first attempt: use stack to reverse linked list

def reverse(head):
    
    node_stack = []
    
    while head:
        
        node_stack.append(head)
        head = head.nextnode
    
    head = node_stack.pop()
    
    while node_stack:
        head.nextnode = node_stack.pop()
        head = head.nextnode
    
    head.nextnode = None
```

```python
# one traverse and reverse in place

def reverse(head):
    
    current = head
    prev = None
    next_node = None
    
    while current:
        next_node = current.nextnode    # copy next_node is important before assign current.nextnode = prev
        current.nextnode = prev
        prev = current
        current = next_node
    
    return prev  # prev is the original tail and new head for the reversed linked list
```
