# Sentence Reversal

## Problem

Given a string of words, reverse all the words. For example:

Given:
    
    'This is the best'

Return:

    'best the is This'

As part of this exercise you should remove all leading and trailing whitespace. So that inputs such as:

    '  space here'  and 'space here      '

both become:

    'here space'

## Solution

Fill out your solution below:

```python
def rev_word1(s):
    
    if len(s) == 0:
        return ""
    
    s1 = s.split(" ")
    s2 = [s1[i].strip() for i in range(len(s1)-1, -1, -1) if s1[i] !=""]
    
    return " ".join(s2)
```

```python
def rev_word2(s):
    return " ".join(s.split()[::-1])  # split(): default separator is any whitespace
```

Basically we shouldn't use the above solutions, specifically that "split()" python function, in a real interview. We could mention those solutions at first though. Then we should implement the split() function by ourselves. Here is one solution:

```python
def rev_word(s):
    
    length = len(s)
    output = []
    i = 0
    
    while i < length:
        
        if s[i] != " ":
            word_start = i
            
            # use this while loop to find each word without whitespace
            while i < length and s[i] != " ":
                i += 1
                
            output.append(s[word_start:i])
        
        i += 1
    
    return " ".join(output[::-1])  # reverse the list
```

```python
rev_word('Hi John,   are you ready to go?')
```

```python
rev_word('    space before')
```
