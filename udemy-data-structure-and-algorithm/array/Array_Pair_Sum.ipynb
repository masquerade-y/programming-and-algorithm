# Array Pair Sum

## Problem

Given an integer array, output all the ** *unique* ** pairs that sum up to a specific value **k**.

So the input:
    
    pair_sum([1,3,2,2],4)

would return **2** pairs:

     (1,3)
     (2,2)

**NOTE: FOR TESTING PURPOSES CHANGE YOUR FUNCTION SO IT OUTPUTS THE NUMBER OF PAIRS**

## Solution

Fill out your solution below:

```python
def pair_sum2(arr,k):
    
    idx_map = {}
    pairs = set()
    
    for i, num in enumerate(arr):
        if k-num in idx_map:
            new_pair = (num, k-num)
            pairs.add(tuple(sorted(new_pair)))
        else:
            idx_map[num] = i
    
    return(len(pairs))
```

```python
def pair_sum(arr,k):
    
    # check up edge case: less than 2 elements in arr
    if len(arr) < 2:
        return print('Warning: less than 2 elements in arr')
    
    seen = set()
    pairs = set()
    
    for num in arr:
        if k-num in seen:  # both search and insert operation of a set are O(1) since set is implemented based on hash table
            new_pair = (num, k-num)
            pairs.add((min(new_pair), max(new_pair)))
        else:
            seen.add(num)
    
    return(len(pairs))
```
