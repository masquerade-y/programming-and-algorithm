# String Compression

## Problem

Given a string in the form 'AAAABBBBCCCCCDDEEEE' compress it to become 'A4B4C5D2E4'. For this problem, you can falsely "compress" strings of single or double letters. For instance, it is okay for 'AAB' to return 'A2B1' even though this technically takes more space. 

The function should also be case sensitive, so that a string 'AAAaaa' returns 'A3a3'.

## Solution
Fill out your solution below:

```python
def compress1(s):
    
    length = len(s)
    i = 0
    output = ""
    
    while i < length:
        letter = s[i]
        counter = 0
        
        while i < length and s[i] == letter:
            counter += 1
            i += 1
        output = output + letter + str(counter)
    
    return output
```

```python
compress1('AAAAABBBBCCCC')
```

```python
def compress(s):
    
    # edge case
    if len(s) == 0:
        return ""
    elif len(s) == 1:
        return s + "1"
    
    output = ""
    i = 1
    counter = 1
    
    while i < len(s):
        
        if s[i] == s[i-1]:
            counter += 1
        else:
            output = output + s[i-1] + str(counter)
            counter = 1
            
        i += 1
    
    output = output + s[i-1] + str(counter)  # add the last letter and counter
        
    return output
```

```python
compress('AABBCcc')
```
