# Unique Characters in String

## Problem
Given a string,determine if it is compreised of all unique characters. For example, the string 'abcde' has all unique characters and should return True. The string 'aabcde' contains duplicate characters and should return false.

## Solution
Fill out your solution below:

```python
# use set() built-in characteristics
def uni_char1(s):
    return len(set(s))==len(s)
```

```python
# use a dictionary
def uni_char(s):
    
    # edge case
    if len(s) == 0:
        return True
    
    counter = {} #use set() is better, no need a dictionary
    
    for c in s:
        if c in counter:
            return False
        else:
            counter[c] = 1
    
    return True
```

```python
# use set(), set search time is O(1)
def uni_char(s):
    
    # edge case
    if len(s) == 0:
        return True
    
    seen = set()
    
    for c in s:
        if c in seen:
            return False
        else:
            seen.add(c)
    
    return True
```
