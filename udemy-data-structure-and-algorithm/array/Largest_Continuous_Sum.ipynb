# Largest Continuous Sum

## Problem
Given an array of integers (positive and negative) find the largest continuous sum. 

## Solution

Fill out your solution below:

```python
def large_cont_sum(arr): 
    
    # edge case: empty array
    if len(arr) == 0:
        return 'Empty array!'
    
    # define current sum and max sum and set to arr[0]
    # current sum not set to 0 because there could be all negative numbers in the array
    max_sum = current_sum = arr[0]
    
    for num in arr[1:]:
        # if current_sum is negative, start new current_sum with num
        current_sum = max(current_sum + num, num)
        
        max_sum = max(max_sum, current_sum)
        
    return max_sum
```

```python
large_cont_sum([1,2,-1,3,4,10,10,-10,-1])
```

```python
large_cont_sum([-9,-2,-10,-1])
```
